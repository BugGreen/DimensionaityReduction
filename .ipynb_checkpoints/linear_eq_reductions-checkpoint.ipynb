```python
from dim_reduction_methods import apply_dimensionality_reduction, is_linear_combination
from visualizations import plot_reduced_data, generate_points_3vars
import numpy as np
```

We will start with the following system of linear equations:
$3x + 4y     = 22$
$ x +  y +  z = 10$
$   -  y +  z = 12$

And the equation:
$3y + z = -16$

Then, we will apply a series of dimensionality reduction methods to them. After that, we will visualize the reduced data to find if it can be determined if the equation is a linear combination of the system of linear equations by just looking at the visualizations.

First, let's determine if the single equation is indeed a linear combination of the system of linear equations:

```python
system_coefficients = [
    [3, 4, 0],  # 3x + 4y + 0z
    [1, 1, 1],  # 1x + 1y + 1z
    [0, -2, 1]  # 0x - 2y + 1z
]
constants = [22, 10, 12]  # Constants for the system equations
given_equation = [0, 3, 1]  # Coefficients for 0x + 3y + 1z
given_constant = -16  # Constant for the given equation

# Check if the given equation is a linear combination of the system
result = is_linear_combination(given_equation, system_coefficients, constants, given_constant)
print("Is the given equation a linear combination of the system? ", "Yes" if result else "No")
```

```python
# Create points for each system equation and the given equation
system_eqs = [
    [3, 4, 0, 22],   # 3x + 4y + 0z = 22
    [1, 1, 1, 10],   # x + y + z = 10
    [0, -2, 1, 12]   # -2y + z = 12
]
given_eq = [0, 3, 1, -16]  # 0x + 3y + 1z = -16

# Generate points
all_points = np.vstack([generate_points_3vars(eq) for eq in system_eqs + [given_eq]])
labels = ['system'] * 300 + ['given'] * 100  # Adjust counts based on number of points per equation

# Apply dimensionality reduction
reduced_data_pca = apply_dimensionality_reduction(all_points.copy(), method='pca')
reduced_data_tsne = apply_dimensionality_reduction(all_points.copy(), method='tsne')
reduced_data_umap = apply_dimensionality_reduction(all_points.copy(), method='umap')

plot_reduced_data(reduced_data_pca, labels, 'PCA')
plot_reduced_data(reduced_data_tsne, labels, 'TSNE')
plot_reduced_data(reduced_data_umap, labels, 'UMAP')
```
